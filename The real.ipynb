```python
import pandas as pd
import numpy as np
missing_values = ["배달전문업체","심부름"]
df = pd.read_csv("seoul_delivery.csv",na_values = missing_values)
df = df.filter(['업종','시군구','통화건수'])
df_grouped=df.groupby(['업종','시군구'])
for a,b in df_grouped:
    print(a)
    print(b)
```

```python
GDP_seoul = pd.read_csv('report.txt',sep='\t')
GDP_seoul=GDP_seoul.filter(['자치구','인구수','월소득'])

GDP_seoul
```

```python
China=df.query('업종.str.contains("중국")')
```

```python
bb = China.groupby('시군구')
bb =bb['통화건수'].sum()
bb
```

```python
bb.index
```

```python
temp2 = GDP_seoul.groupby('자치구')
temp2 = temp2['인구수'].sum()
temp2
```

```python
temptable = pd.concat([bb, temp2], axis=1)

temptable['1인당 배달비율']=temptable['통화건수'].astype('float')/temptable['인구수'].astype('float')
```

```python
temptable
```

```python
temp3 = GDP_seoul.groupby('자치구')
temp3 = temp3['월소득'].sum()
temp3
```

```python
temptable = pd.concat([temptable, temp3], axis=1)
```

```python
temptable
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temptable=temptable.sort_values(by=['월소득'], axis=0)
temptable
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
plt.figure(figsize=(20,10))
plt.plot(temptable.index,temptable['1인당 배달비율'])
plt.show()
```

```python
ck=df.query('업종.str.contains("보쌈")')
cc = ck.groupby('시군구')
cc =cc['통화건수'].sum()
t2 = GDP_seoul.groupby('자치구')
t2 = t2['인구수'].sum()
t2table = pd.concat([cc, t2], axis=1)
t2table['1인당 배달비율']=t2table['통화건수'].astype('float')/t2table['인구수'].astype('float')
t3 = GDP_seoul.groupby('자치구')
t3 = t3['월소득'].sum()
t2table = pd.concat([t2table, t3], axis=1)
t2table=t2table.sort_values(by=['월소득'], axis=0)
plt.figure(figsize=(20,10))
plt.plot(t2table.index,t2table['1인당 배달비율'])
plt.show()
```
